```python
from IPython.display import HTML, display
import re

def csv_line_as_table_row(line):
    prefix = '<tr><td>'
    postfix = '</td></tr>'
    
    row = line[1:-2].replace('",,"','</td><td></td><td>').replace('","','</td><td>').replace('"""','').replace('""', '"').replace('",</td>', '</td>')
    
    return prefix + row + postfix

html_table = '<table>'

with open('rawdata.csv') as f:
    for l in f:
        html_table = html_table + csv_line_as_table_row(l)
html_table = html_table + '</table>'
display(HTML(html_table))
```
